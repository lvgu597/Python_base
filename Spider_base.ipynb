### 2. 常见反爬方法
#### 2.1 构造herders
-  User-Agent 
   + User-Agent，用户代理。让服务器能够识别用户浏览器信息。比较常见的是设置随机User-Agent，设置User-Agent有两个作用：1. 通过服务器的验证 2. 认为服务器错认为是不同的用户在请求页面，从而降低被识别成robot的几率。虽然简单，但是很有效。
-  Referer 
   + Referer的作用是告诉服务器我是从哪一个页面链接过来的，对于一些详情页的抓取比较有效。
-  Authorization 
   + 这个字段是用来进行身份验证的，一般表示用户的唯一标识，如果有，请务必加上
-  X-Requested-With 
   + 这个字段表示页面请求为同步还是Ajax，比较常见的为X-Requested-With ：XMLHttpRequest，如果你发现请求的接口数据跳到网页了，那么就是这个字段的锅。
-  cookie
   + cookie是网站为了辨别用户身份、进行 session 跟踪而储存在用户本地终端上的数据 ，在某些需要身份验证的网站，加上这个字段可以有效的避开登录，但现在越来越多的网站cookie 都是有时效性的。
