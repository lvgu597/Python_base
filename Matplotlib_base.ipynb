### 1. 基本名词解释
+ Figure结构
    + 
+ Figutre：画布
    + `fig = plt.figure()`:创建一个画布，然后在画布上添加各种元素
+ Axes
    + `ax = fig.add_subplot(1, 1, 1)`
    + 可以理解为你要放到画布Figure上的各种东西，比如你要画几张图就是几个axes
    + Figure上只有一张图就是一个axes，有多个subplot，那么每一个subplot就是一个axes
+ Axis
    + `ax.xaxis/ax.yaxis`: xy坐标轴
    + 每个坐标轴实际上也是由竖线和数字组成，每一个竖线其实也是一个axis的subplot，因此`ax.xaxis`也存在`axes`这个对象。对axes这个对象进行修改就回修改该xaxis在Figure上的表现

### 2. 图像的各个部位名词解释
+ 图像内部结构
    + 
+ 基本操作：
    + 创建画布：
        + `fig, ax = plt.subplots(numrows, numcols)`创建一个画布，将其赋值给fig，同时在画布上创建一axes，赋值给ax,`numrows表示行`,`numcol表示列`
        + `ax1 = plt.subplot(numrows, numcols, plot_number)`在默认画布figure(1) 上创建一个axes,`numrows表示行`,`numcol表示列`,`plot_number表示该axes位置`
            + fig.xxx 为对画布的操作；ax.xxx为对axes的操作
    + 添加数据：
        + `ax.plot(A, B)`
    + 添加细节
        + 设置title、横纵坐标标题、横纵坐标属性等都是在axes上进行
        + 设置横纵坐标、画图相关在axes.xaxis或axes.yaxis上进行

```python
## 画图基本操作
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,7)) #创建一个画布，一个axes

#fig, ax = plt.subplots(2,1,figsize=(10,7)) #创建一个画布，多个axes，ax为list

# ax1 = plt.subplot(2,1,1)   #用subplot在就近画布figure(1)上创建一个axes，其位置为2行1列第1个
# plt.plot([1,2,3], [1,2,3])
# ax2 = plt.subplot(2,1,2)  #用subplot在就近画布figure(1)上创建一个axes，其位置为2行1列第2个
# plt.plot([1,2,3], [1,2,3])
```

```python
##　添加数据
import numpy as np
A = np.arange(5)
B = A**2 # A^2
fig, ax = plt.subplots(figsize=(10,7)) #创建一个画布，一个axes
ax.plot(A, B)
ax.plot(B, A)

##设置细节
ax.set_title('Title', fontsize = 30)
ax.set_xlabel('x_lable')
ax.set_ylabel('y_lable')
ax.xaxis.set_tick_params(rotation=45,labelsize=18,colors='w') 
start, end = ax.get_xlim() 
ax.xaxis.set_ticks(np.arange(start, end,1)) 
ax.yaxis.tick_right()
```
